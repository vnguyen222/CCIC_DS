```python
import pandas as pd
import re
```

```python
# Read .csv data
data = pd.read_csv("RickAndMortyScripts.csv")
# Drop unnecessary data
data.drop(columns=["index", "season no.", "episode no.", "episode name"], inplace=True)
data
```

```python
# Cleaning the data
for index, row in data.iterrows():
    # Remove all of the symbols and numbers 
    line = row["line"]
    line = re.sub(r'[^\w\s]', '', line)
    line = line.lower()
    data.at[index, "line"] = line

data
```
