```python
import requests
import pandas as pd
```

```python
# Request data from TMDB database with their API

def get_toprated_data():
    page_count = 1
    entries = []

    while True:
        url = "https://api.themoviedb.org/3/movie/top_rated?api_key=87983fbe976988a9e22330ebf030a1ae&page={0}".format(page_count)
        req = requests.get(url)
        if not req.ok:
            break

        entries = entries + req.json()["results"]
        page_count += 1
    
    return entries
```

```python
MOVIES = get_toprated_data()

# Temp Code
# url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=87983fbe976988a9e22330ebf030a1ae&page=1'
# req = requests.get(url)
# movies = req.json()["results"]
```

# All Revenue
- Highest revenue for a single movie
- Highest average revenue across all movies

```python
data = pd.DataFrame(MOVIES)
data.head()
```

```python
data.drop(columns=["adult", "backdrop_path", "original_language", "original_title","overview", "poster_path", "release_date", "video", "vote_average", "vote_count"], inplace=True)
data
```

```python
# Get Revenue 
def get_revenue():
    revenues = []

    for index, movie_entry_row in data.iterrows():
        url = 'https://api.themoviedb.org/3/movie/{0}?api_key=87983fbe976988a9e22330ebf030a1ae'.format(movie_entry_row["id"])
        req = requests.get(url)
        
        temp = req.json()["revenue"]
        revenues.append(temp)
        # print(temp)

    return revenues

revenues_list = get_revenue()
revenues_list
```

```python
REVENUES_COL = pd.DataFrame(revenues_list)
data = pd.concat([data, REVENUES_COL], axis = 1)
data.rename(columns = {0 : "revenue"}, inplace = True)
data.head()
```

```python
data.sort_values("revenue", ascending=False, inplace=True)
data.drop(data.loc[data['revenue']==0].index, inplace=True)
data.tail()
```

```python
data.to_csv("moviedata.csv")
```
