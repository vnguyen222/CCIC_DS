# Movies Directors
- Top 10 movies directors in a single movie 

 Who are the ten directors with the highest revenue for a single movie?

```python
import pandas as pd
import requests
```

```python
data = pd.read_csv("moviedata.csv")
data.drop(columns=["oldid", "genre_ids", "popularity"], inplace=True)
data.head(20)
```

```python
top20movies = data.head(20)

def get_top10_directors():
    directors = []

    for index, movie_entry_row in top20movies.iterrows():
        url = 'https://api.themoviedb.org/3/movie/{0}/credits?api_key=87983fbe976988a9e22330ebf030a1ae'.format(movie_entry_row["id"])
        req = requests.get(url)

        single_movie_directors = []

        crew_list = req.json()["crew"]
        for crew_member in crew_list:
            if crew_member["job"] == 'Director':
                single_movie_directors.append(crew_member["name"])
        directors.append(single_movie_directors)
    
    return directors


TOP10_DIRECTORS = get_top10_directors()
TOP10_DIRECTORS
```

```python
director_list = TOP10_DIRECTORS.copy()

for i in range(0, len(director_list)):
    if len(director_list[i]) > 1:
        director_list[i] = [director_list[i][0] + ", " + director_list[i][1]]

director_df = pd.DataFrame(director_list)

top20movies = pd.concat([top20movies, director_df], axis = 1, sort=False)
top20movies.rename(columns = {0 : "director"}, inplace = True)
top20movies.drop_duplicates(subset=["director"], keep="first", inplace=True)
top20movies
```

```python
top20movies.drop(columns=["id"], inplace=True)
toptendirectors_singlemovie = top20movies.head(10)
toptendirectors_singlemovie
```
