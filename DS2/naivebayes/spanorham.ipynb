```python
import pandas as pd
import string
import ite
```

```python
# Get data
data = pd.read_csv("spam.csv", encoding="UTF-8")

# Get list of stop words
stopwords = []
with open("stopwords.txt", "r") as stopwords_file:
    filecontents = stopwords_file.read()
    stopwords = filecontents.split("\n")
```

```python
data.drop(data.columns[[2,3,4]], axis=1, inplace=True)
data
```

```python
data["Label"].value_counts()["spam"]
```

```python
data_only_spam = data.loc[data["Label"] == "spam"]
data_only_spam
```

```python
words_occurances = {}
for index, row in data_only_spam.iterrows():
    message = row["Message"].lower()
    # message = message.tran
    message_words = message.split(" ")
    for word in message_words:
        if word not in words_occurances:
            words_occurances[word] = 1
        else:
            words_occurances[word] += 1
        
words_occurances
sorted_word_occurances = dict(sorted(words_occurances.items(), key=lambda item: item[1], reverse=True))
sorted_word_occurances
```

```python
# Get common words in spam that are not stop words
spam_stop_words = {}
for key, value in sorted_word_occurances.items():
    if key not in stopwords:
        spam_stop_words[key] = value
    
spam_stop_words
```
