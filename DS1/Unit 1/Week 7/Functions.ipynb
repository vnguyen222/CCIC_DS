```python
from time import time
from random import randint
```

```python
def sort(li):
    if len(li) == 0:
        return []
    
    sorted_li = []

    for i in range(1,len(li)):
        current_value = li[i]
        possible_index = i

        while possible_index > 0 and current_value < li[possible_index-1]:
            li[possible_index] = li[possible_index-1]
            possible_index -= 1
        
        li[possible_index] = current_value
    
    return li

def generate_list(length):
    generated_list = []
    for x in range(length):
        generated_list.append(randint(0, 200))
    return generated_list

performance = []
max_list_length = 100

for iter in range(max_list_length):
    start = time()
    sort(generate_list(iter))
    end = time()

    execution_time = end - start

    performance.append([iter, execution_time])

print(performance)
```

```python
from matplotlib import pyplot as plt

plt.plot(performance)
```
